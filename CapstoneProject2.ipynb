```python
# 1. store_and_fwd_flag 
# handling missing value dengan modus karena tipe data object

modus_1 = df['store_and_fwd_flag'].mode()[0]
df['store_and_fwd_flag'].fillna(modus_1, inplace=True)
```

```python
#2. RatecodeID
#cek normalitas, normal = mean ; tidak normal = median
from scipy.stats import shapiro
from scipy.stats import normaltest

stat, pvalue = shapiro(df['RatecodeID'])
if pvalue > 0.05:
    print(f'pvalue={pvalue:.5f}, kita mengasumsikan data terdistribusi normal')
else:
        print(f'pvalue={pvalue:.5f}, kita mengasumsikan data TIDAK terdistribusi normal')
```

```python
#mengisi NaN dengan nilai mean karena kolom ratecode terdirtribusi normal
mean_ratecodeid = df['RatecodeID'].mean()
df['RatecodeID'] = df['RatecodeID'].fillna(mean_ratecodeid)
```

```python
df['RatecodeID'] = df['RatecodeID'].astype(int)
```

```python
# menghapus baris yang 'RatecodeID' nya == 99
df = df.drop(df[df.RatecodeID == 99].index)
```

```python
df['RatecodeID'].unique()
```

```python
stat, pvalue = shapiro(df['passenger_count'])
if pvalue > 0.05:
    print(f'pvalue={pvalue:.5f}, kita mengasumsikan data terdistribusi normal')
else:
        print(f'pvalue={pvalue:.5f}, kita mengasumsikan data TIDAK terdistribusi normal')
```

```python
#mengisi NaN dengan nilai mean karena kolom passenger terdirtribusi normal
mean_pass = df['passenger_count'].mean()
df['passenger_count'] = df['passenger_count'].fillna(mean_pass)

df['passenger_count'] = df['passenger_count'].astype(int)

df['passenger_count'].unique()
```

```python
df.drop(df[df['passenger_count'] == 0].index, inplace=True)
df[df['passenger_count']==0]
```

```python
#4. Drop ehail fee 
df.drop('ehail_fee', axis=1, inplace=True)
```

```python
df.head()
```

```python
#handling negative value dengan menjadikan nilai negatif ke absolute
df['fare_amount'] = df['fare_amount'].abs()
df['extra'] = df['extra'].abs()
df['mta_tax'] = df['mta_tax'].abs()
df['tip_amount'] = df['tip_amount'].abs()
df['improvement_surcharge'] = df['improvement_surcharge'].abs()
df['total_amount'] = round(df['total_amount'].abs(),1)
df['congestion_surcharge'] = df['congestion_surcharge'].abs()
```

```python
color = sns.color_palette("Paired")
scale = 'coolwarm_r'

cols = ['pickup_day', 'RatecodeID', 'PUBorough', 'payment_type', 'trip_type']

fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(18,9))
axs = axs.flatten()

for i, col in enumerate(cols):
    bars = sns.countplot(data=df, x=col, palette=color, ax=axs[i])
    axs[i].set_title(col)
    for idx, bar in enumerate(bars.patches):
        x_value = bar.get_x() + bar.get_width() / 2
        y_value = bar.get_height()
        label = "{:.0f}".format(bar.get_height())
        axs[i].text(x_value, y_value, label, ha='center', va='bottom') 

plt.tight_layout(w_pad=0.5, h_pad=2)
plt.show()
```

```python
count_by_pickuphour = df[["passenger_count", "pickup_hour"]].groupby("pickup_hour").sum().sort_values("passenger_count", ascending = False).reset_index()
count_by_pickuphour
```

```python
plt.figure(figsize=(18, 7))
ax = sns.lineplot(y='passenger_count', x='pickup_hour', data=count_by_pickuphour, color='blue', marker='o')
plt.title('Total Penumpang Berdasarkan Jam')

# menambahkan value
ax.set_xticks(count_by_pickuphour['pickup_hour'])
ax.set_xticklabels(count_by_pickuphour['pickup_hour'])

# menambahkan anotasi value
for x, y in zip(count_by_pickuphour['pickup_hour'], count_by_pickuphour['passenger_count']):
    ax.annotate(f"{y}", (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# menambahkan label pada sumbu x, dan sumbu y
plt.xlabel('Jam')
plt.ylabel('Total Penumpang')

plt.show()
```

```python
# Create a column indicating whether the day is a weekday or weekend
df['day'] = df['pickup_day_no'].apply(lambda x: 'weekday' if x < 5 else 'weekend')

# Group by 'pickup_hour' and 'day_type' and count occurrences
hourly_counts = df.groupby(['pickup_hour', 'day']).size().reset_index(name='ride_count')

# Separate data for weekdays and weekends
weekday = hourly_counts[hourly_counts['day'] == 'weekday']
weekend = hourly_counts[hourly_counts['day'] == 'weekend']

# Plot the peak hours for weekdays
plt.figure(figsize=(12, 6))
plt.plot(weekday['pickup_hour'], weekday['ride_count'], label='Weekday', marker='o')
plt.title('Peak Hours for Weekday Commute')
plt.xlabel('Hour of the Day')
plt.ylabel('Ride Count')
plt.legend()
plt.grid(True)
plt.xticks(range(24))  # Show all hours on the x-axis
plt.savefig('weekday_peak_hours.png', transparent=True)  # Save with transparent background
plt.show()

# Plot the peak hours for weekends
plt.figure(figsize=(12, 6))
plt.plot(weekend['pickup_hour'], weekend['ride_count'], label='Weekend', marker='o', color='orange')
plt.title('Peak Hours for Weekend Commute')
plt.xlabel('Hour of the Day')
plt.ylabel('Ride Count')
plt.legend()
plt.grid(True)
plt.xticks(range(24))  # Show all hours on the x-axis
plt.savefig('weekend_peak_hours.png', transparent=True)  # Save with transparent background
plt.show()
```
